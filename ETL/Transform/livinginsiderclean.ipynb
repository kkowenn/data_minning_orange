```python
import pandas as pd
import random

# Load the Zones and Districts CSV file
zones_districts_df = pd.read_csv("zones_districts.csv")
zones_districts = dict(zip(zones_districts_df["Zone"], zones_districts_df["District"]))

# Load the main dataset
df = pd.read_csv("unclean.csv")

# Clean and convert columns
df["price"] = pd.to_numeric(df["price"].str.replace(r"[^\d]", "", regex=True), errors='coerce')  # Remove non-numeric characters and convert to float
df["price_per_space"] = pd.to_numeric(df["price_per_space"].str.replace(r"[^\d]", "", regex=True), errors='coerce')  # Remove non-numeric characters and convert to float
df["space"] = pd.to_numeric(df["space"].str.replace(r"Sq.m.", "", regex=True).str.strip(), errors='coerce')  # Remove "Sq.m." and convert to float
df["floor"] = pd.to_numeric(df["floor"].str.extract(r"(\d+)")[0], errors='coerce')  # Extract the first floor number and convert to int
df["bedroom"] = pd.to_numeric(df["bedroom"].str.replace(r"[^\d]", "", regex=True), errors='coerce')  # Keep numeric characters and convert to int
df["bathroom"] = pd.to_numeric(df["bathroom"].str.replace(r"[^\d]", "", regex=True), errors='coerce')  # Keep numeric characters and convert to int

# Function to convert "K" and "M" to numeric values
def convert_clicks(clicks):
    if isinstance(clicks, str):
        if 'K' in clicks.upper():  # Ensure case-insensitivity
            return float(clicks.upper().replace('K', '')) * 1000
        elif 'M' in clicks.upper():  # Ensure case-insensitivity
            return float(clicks.upper().replace('M', '')) * 1000000
    return clicks

# Apply the conversion function to the 'click' column
df['click'] = df['click'].apply(convert_clicks)

# Drop rows with NaN values
df.dropna(subset=["price", "price_per_space", "space", "floor", "bedroom", "bathroom"], inplace=True)

# Rename the "location" column to "detail location"
df.rename(columns={"location": "detail location"}, inplace=True)

# Create Zone and District columns
df["Zone"] = ""
df["District"] = ""

# Populate Zone and District columns based on detail location
for idx, row in df.iterrows():
    detail_location = row["detail location"].lower()
    matched = False
    for zone, district in zones_districts.items():
        if zone.lower() in detail_location:
            df.at[idx, "Zone"] = zone
            df.at[idx, "District"] = district
            matched = True
            break
    if not matched:
        locations = [loc.strip() for loc in row["detail location"].split(",")]
        df.at[idx, "Zone"] = random.choice(locations)
        df.at[idx, "District"] = random.choice(locations)

# Load the nearby data
nearby_df = pd.read_csv("nearby.csv")

# Ensure NearBy and Distance columns are treated as strings and case-insensitive
nearby_df["NearBy"] = nearby_df["NearBy"].str.lower()

# Clean the Distance column and handle "Not Found"
nearby_df["Distance"] = nearby_df["Distance"].str.replace(" Km.", "", regex=False)
nearby_df["Distance"] = pd.to_numeric(nearby_df["Distance"], errors='coerce')  # Convert to numeric, setting invalid entries to NaN

# Handle NaN values in Distance (either drop or assign default value)
nearby_df.dropna(subset=["Distance"], inplace=True)  # Option 1: Drop rows with NaN
# nearby_df["Distance"].fillna(99999, inplace=True)  # Option 2: Replace NaN with a default value

# Define keywords
keywords = {
    "BTS": "bts",
    "Airport Link": "airport link",
    "Hospital": "hospital",
    "Education place": ["collage", "institute", "university"],  # New keyword group for educational places
    "Lotus": "lotus",
    "Suvarnabhumi Airport": "suvarnabhumi airport",
    "Robinson": "robinson",
    "Big C": "big c",
    "Makro": "makro",
    "Don Mueang International Airport": "don mueang international airport",
    "MRT": "mrt",
    "Mall": ["central", "mall", "plaza"]  # New keyword for malls
}

# Create columns for keywords in df
for key in keywords:
    df[key] = 0

# Check for keywords in nearby_df and update df
for idx, row in df.iterrows():
    link = row["link"]
    if link in nearby_df["link"].values:
        nearby_rows = nearby_df[nearby_df["link"] == link]
        for _, nearby_row in nearby_rows.iterrows():
            for key, values in keywords.items():
                if isinstance(values, list):
                    if any(value.lower() in nearby_row["NearBy"] for value in values):
                        df.at[idx, key] = 1
                else:
                    if values.lower() in nearby_row["NearBy"]:
                        df.at[idx, key] = 1

# Create new columns for nearest train station information
df["Nearest Train station name"] = ""
df["Nearest Train station type"] = ""
df["Nearest Train station name distance"] = ""

# Populate nearest train station information
for idx, row in df.iterrows():
    link = row["link"]
    if link in nearby_df["link"].values:
        nearby_rows = nearby_df[nearby_df["link"] == link]
        train_stations = nearby_rows[nearby_rows["NearBy"].str.contains("bts|mrt|airport link")]
        if not train_stations.empty:
            nearest_station = train_stations.loc[train_stations["Distance"].idxmin()]
            df.at[idx, "Nearest Train station name"] = nearest_station["NearBy"]
            df.at[idx, "Nearest Train station name distance"] = nearest_station["Distance"]
            if "bts" in nearest_station["NearBy"]:
                df.at[idx, "Nearest Train station type"] = "BTS"
            elif "mrt" in nearest_station["NearBy"]:
                df.at[idx, "Nearest Train station type"] = "MRT"
            elif "airport link" in nearest_station["NearBy"]:
                df.at[idx, "Nearest Train station type"] = "Airport Link"

# Remove rows with "Not Found" in any columns
df.replace("Not Found", pd.NA, inplace=True)
df.dropna(inplace=True)

# Save the updated data to a new CSV file with the specified format
final_columns = [
    "link", "post_title", "condo_name", "detail location", "price", "price_per_space", "space",
    "floor", "bedroom", "bathroom", "click", "Page", "Zone", "District",
    "BTS", "MRT", "Airport Link", "Hospital", "Education place", "Mall",
    "Lotus", "Robinson", "Big C", "Makro", "Don Mueang International Airport",
    "Suvarnabhumi Airport", "Nearest Train station name", "Nearest Train station type", "Nearest Train station name distance"
]
df[final_columns].to_csv("updated_cleaned_data2.csv", index=False)

# Display the updated data for verification
print(df[final_columns])
```
