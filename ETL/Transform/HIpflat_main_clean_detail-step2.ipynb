```python
import pandas as pd

# Read the CSV file, skipping bad lines
df = pd.read_csv("converted_prices.csv")

# Check the first few rows of the dataframe
df
```

```python
# List of facilities
facilities = [
    "Gym", "Garden", "Security", "Sauna", "Children area", "Swimming pool",
    "Library", "Jacuzzi", "Balcony", "Grill", "Parking", "Tennis court"
]

# Step 1: Add columns for each facility
for facility in facilities:
    # Step 2: Check if each facility is present in any column of the row
    df[facility] = df.apply(
        lambda row: 1 if any(facility in str(value) for value in row) else 0,
        axis=1
    )

# Display the updated DataFrame
df.head()
```

```python
# Save the updated DataFrame to an Excel file
df.to_csv('draft.csv', index=False)

print("The DataFrame has been converted and saved to 'draft.csv'")
```

```python
# Function to extract numeric values from evolution strings
def extract_percentage(value):
    if isinstance(value, str):  # Check if the value is a string
        return float(value.split("%")[0].replace("+", "").strip())
    return None  # Return None for NaN or invalid values

# Rename columns
df.rename(
    columns={
        "for_rent_evolution": "Rent percentage evolution compared to December 2023",
        "for_sale_evolution": "Sale percentage evolution compared to December 2023"
    },
    inplace=True
)

# Apply the function to transform the values
df["Rent percentage evolution compared to December 2023"] = df[
    "Rent percentage evolution compared to December 2023"
].apply(extract_percentage)

df["Sale percentage evolution compared to December 2023"] = df[
    "Sale percentage evolution compared to December 2023"
].apply(extract_percentage)

# Display the updated DataFrame
print(df)
```

```python
# Save the updated DataFrame to an Excel file
df.to_csv('draf2.csv', index=False)

print("The DataFrame has been converted and saved to 'draft2.csv'")
```

```python
# List of pets-related phrases and their mapping
pets_list = {
    "Pets: All allowed All allowed": "Allowed all",
    "Pets: Cats and dogs small Cats and dogs small": "Allowed small cats and dogs",
    "Pets: Cats only Cats only": "Allowed cats only",
    "Pets: Dogs only Dogs only": "Allowed dogs only",
    "Pets: Cats and dogs any size Cats and dogs any size": "Allowed any size cats and dogs"
}

# Step 1: Add the "Pets allowance" column
def check_pets(row):
    for phrase, result in pets_list.items():
        if any(phrase in str(value) for value in row):
            return result
    return "Not allowed"

df3 = df
# Apply the function to each row
df3["Pets allowance"] = df.apply(check_pets, axis=1)

# Display the updated DataFrame
print(df3)
```

```python
# Save the updated DataFrame to an Excel file
df3.to_csv('draf3.csv', index=False)

print("The DataFrame has been converted and saved to 'draft3.csv'")
```

```python
import re

df4 = df3

# Function to extract the required information
def extract_common_area_fee(row):
    for value in row:
        match = re.search(r"Common area management fee: ฿(\d+)/sqm", str(value))
        if match:
            return int(match.group(1))
    return ""

def extract_management_company(row):
    for value in row:
        match = re.search(r"Common area management company: ([\w\s-]+)", str(value))
        if match:
            return match.group(1).strip()
    return "Not provided"

def extract_sinking_fund(row):
    for value in row:
        match = re.search(r"Sinking fund: ฿(\d+)/sqm", str(value))
        if match:
            return int(match.group(1))
    return ""

def extract_parking_ratio(row):
    for value in row:
        match = re.search(r"Ratio of parking spaces to # of units: (\d+)%", str(value))
        if match:
            return int(match.group(1))
    return ""

def extract_parking_type(row):
    for value in row:
        match = re.search(r"Parking type: ([\w\s]+)", str(value))
        if match:
            return match.group(1).strip()
    return "Not provided"

# Step 2-6: Apply the functions to extract values
df4["Common area management fee in baht"] = df.apply(extract_common_area_fee, axis=1)
df4["Common area management company"] = df.apply(extract_management_company, axis=1)
df4["Sinking fund per sqm"] = df.apply(extract_sinking_fund, axis=1)
df4["Ratio of parking spaces"] = df.apply(extract_parking_ratio, axis=1)
df4["Parking types"] = df.apply(extract_parking_type, axis=1)
```

```python
# Save the updated DataFrame to an Excel file
df4.to_csv('draf4.csv', index=False)

print("The DataFrame has been converted and saved to 'draft4.csv'")
```

```python
df5 = df4

import pandas as pd
import re

df5 = df4

# Step 1: Add the new columns
df5["date the condo was completed"] = ""
df5["year the condo was completed"] = ""

# Function to extract the date and year
def extract_date_and_year(row):
    date_match = None
    for value in row:
        # Find dates in the format "Month-Year", "Month Year", or "Month-Year (short years)" like "Jan-95"
        match = re.search(r"(?i)(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[-\s](\d{2}|\d{4})", str(value))
        if match:
            date_match = match.group(0)  # Full date match
            if date_match.lower() == "december 2023":  # Skip "December 2023"
                continue
            break

    # Extract and validate the year from the matched date
    if date_match:
        year_match = re.search(r"\d{2,4}", date_match)  # Match either 2-digit or 4-digit year
        if year_match:
            year = year_match.group(0)
            # Convert 2-digit year to 4-digit (e.g., "95" becomes "1995")
            if len(year) == 2:
                year = f"19{year}" if int(year) > 50 else f"20{year}"
            year = int(year)  # Convert to integer
        else:
            year = None
    else:
        year = None

    return date_match, year

# Apply the function to extract dates and years
df5["date the condo was completed"], df5["year the condo was completed"] = zip(
    *df5.apply(extract_date_and_year, axis=1)
)

# Display the updated DataFrame
print(df5)
```

```python
# Save the updated DataFrame to an Excel file
df5.to_csv('draf5.csv', index=False)

print("The DataFrame has been converted and saved to 'draft5'")
```

```python
df6 = df5

# Regex pattern to extract the structured location
pattern = r"^(.*?),?\s*(Soi\s.*?|.*?Rd\.|.*?Road|.*?),\s*(.*?),\s*(.*?),\s*Bangkok\s*(\d{5}),\s*THA$"

# Function to extract and format the full location
def extract_full_location(location):
    match = re.match(pattern, str(location))
    if match:
        # Combine all components into a single string for "full_location"
        return f"{match.group(1)}, {match.group(2)}, {match.group(3)}, {match.group(4)}, Bangkok {match.group(5)}, THA".strip(", ")
    return None

# Apply the function to create the new "full_location" column
df6["full_location"] = df6["location"].apply(extract_full_location)

# Display the updated DataFrame
print(df6)
```

```python
# Save the updated DataFrame to an Excel file
df6.to_csv('draf6.csv', index=False)

print("The DataFrame has been converted and saved to 'draft6'")
```

```python
df7 = df6

# Regex pattern to extract the structured location
pattern = r"^(.*?),?\s*(Soi\s.*?|.*?Rd\.|.*?Road|.*?),\s*(.*?),\s*(.*?),\s*Bangkok\s*(\d{5}),\s*THA$"

# Function to extract components
def extract_location_components(location):
    match = re.match(pattern, str(location))
    if match:
        return {
            "Building/Place Name": match.group(1).strip(", ") if match.group(1) else None,
            "Soi/Road": match.group(2).strip(", ") if match.group(2) else None,
            "Sub-District": match.group(3).strip(", ") if match.group(3) else None,
            "District": match.group(4).strip(", ") if match.group(4) else None,
            "Postal Code": match.group(5).strip(", ") if match.group(5) else None,
        }
    return {
        "Building/Place Name": None,
        "Soi/Road": None,
        "Sub-District": None,
        "District": None,
        "Postal Code": None,
    }

# Apply the function and create new columns
location_components = df7["location"].apply(extract_location_components)
location_df = pd.DataFrame(location_components.tolist())

# Combine the original DataFrame with the new columns
df7 = pd.concat([df7, location_df], axis=1)

# Display the updated DataFrame
print(df7)
```

```python
# Save the updated DataFrame to an Excel file
df7.to_csv('draf7.csv', index=False)

print("The DataFrame has been converted and saved to 'draft7'")
```
